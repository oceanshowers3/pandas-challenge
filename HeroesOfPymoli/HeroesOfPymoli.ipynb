# Generates a report that breaks down the game's purchasing data into meaningful insights. 
## Includes each of the following:

- Player Count
  - Total Number of Players
- Purchasing Analysis (Total)
  - Number of Unique Items
  - Average Purchase Price
  - Total Number of Purchases
  - Total Revenue
- Gender Demographics
  - Percentage and Count of Male Players
  - Percentage and Count of Female Players
  - Percentage and Count of Other / Non-Disclosed
- Purchasing Analysis (Gender)
  - The below each broken by gender:
    - Purchase Count
    - Average Purchase Price
    - Total Purchase Value
    - Average Purchase Total per Person by Gender
- Age Demographics
  - The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.):
    - Purchase Count
    - Average Purchase Price
    - Total Purchase Value
    - Average Purchase Total per Person by Age Group
- Top Spenders
  - Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
    - SN
    - Purchase Count
    - Average Purchase Price
    - Total Purchase Value
- Most Popular Items
  - Identify the 5 most popular items by purchase count, then list (in a table):
    - Item ID
    - Item Name
    - Purchase Count
    - Item Price
    - Total Purchase Value
- Most Profitable Items
  - Identify the 5 most profitable items by total purchase value, then list (in a table):
    - Item ID
    - Item Name
    - Purchase Count
    - Item Price
    - Total Purchase Value

```python
#import pandas
import pandas as pd
```

```python
#set file path and read CSV file
csvPath = "Resources/purchaseData.csv"
purchaseDf = pd.read_csv(csvPath)
purchaseDf.head()
```

## Player Count
- Total Number of Players

```python
#calculate total (unique) players
totalPlayers = len(pd.unique(purchaseDf['SN']))

#display them in a table
totalPlayersDict = {'Number of Players': [totalPlayers]}
totalPlayersDf = pd.DataFrame(data=totalPlayersDict)
totalPlayersDf
```

## Purchasing Analysis (Total)

- Number of Unique Items
- Average Purchase Price
- Total Number of Purchases
- Total Revenue

```python
#calculate number of unique items
totalCount = len(pd.unique(purchaseDf['Item Name']))

# #calculate total number of purchases
totalPurchases = purchaseDf.count()['Price']

# #calculate total revenue
totalRevenue = purchaseDf.sum()['Price']

# #calculate average purchase price
avgPurchase = round(purchaseDf.mean()['Price'], 2)

#display results of purchasing anaylses in a table
purchaseAnalysisDf = pd.DataFrame({'Number of Unique Items': [totalCount],
                        'Average Purchase Price': avgPurchase,
                        'Total Number of Purchases':totalPurchases,
                        'Total Revenue': totalRevenue})
purchaseAnalysisDf['Average Purchase Price'] = purchaseAnalysisDf['Average Purchase Price'].map('${:,.2f}'.format)
purchaseAnalysisDf['Total Revenue'] = purchaseAnalysisDf['Total Revenue'].map('${:,.2f}'.format)
purchaseAnalysisDf
```

## Gender Demographics
- Percentage and Count of Male Players
- Percentage and Count of Female Players
- Percentage and Count of Other / Non-Disclosed

```python
duplicateDrop = purchaseDf.loc[:,['Gender','SN','Age']]
duplicateDrop = duplicateDrop.drop_duplicates()
totalCount = duplicateDrop['Gender'].value_counts()
genderPerc = totalCount/totalPlayers
genderDf = pd.DataFrame({'Total Count':totalCount,
                         'Percentage of Players':genderPerc})
genderDf['Percentage of Players'] = genderDf['Percentage of Players'].map('{:,.2%}'.format)
genderDf
```

## Purchasing Analysis (Gender)
- The below each broken by gender:
  - Purchase Count
  - Average Purchase Price
  - Total Purchase Value
  - Average Purchase Total per Person by Gender

```python
#calculate the purchase count
purchaseCountGen = purchaseDf.groupby(['Gender']).count()['Price'].rename("Purchase Count")

#calculate the total purchase value
purchaseTotalGen = purchaseDf.groupby(['Gender']).sum()['Price'].rename('Total Purchase Value')

#calculate the average purchase price
purchaseAvgGen = purchaseDf.groupby(['Gender']).mean()['Price'].rename('Average Purchase Price')

#calculate the average purchase total per person by gender
purchaseAvgTotalGen = purchaseTotalGen/genderDf['Total Count']

#display the results in a table 
purchaseGenDf = pd.DataFrame({'Purchase Count': purchaseCountGen,
                             'Average Purchase Price': purchaseAvgGen,
                             'Total Purchase Value': purchaseTotalGen,
                             'Average Purchase Total': purchaseAvgTotalGen
                             
})
purchaseGenDf['Average Purchase Price'] = purchaseGenDf['Average Purchase Price'].map('${:,.2f}'.format)
purchaseGenDf['Total Purchase Value'] = purchaseGenDf['Total Purchase Value'].map('${:,.2f}'.format)
purchaseGenDf['Average Purchase Total'] = purchaseGenDf['Average Purchase Total'].map('${:,.2f}'.format)
purchaseGenDf
```

## Age Demographics
 
- Establish bins for ages

- Categorize the existing players using the age bins

- Calculate the numbers and percentages by age group

- Create a summary data frame to hold the results

```python
#bin those ages!
ageBins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 120]

#label those bins!
ageLabels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#put 'em together now!
duplicateDrop['Age Ranges'] = pd.cut(duplicateDrop['Age'], ageBins, labels=ageLabels)

#now put 'em in the table!
ageCount = duplicateDrop['Age Ranges'].value_counts()
agePerc = ageCount/totalPlayers
ageDf = pd.DataFrame({'Total Count':ageCount,
                         'Percentage of Players':agePerc})
ageDf['Percentage of Players'] = ageDf['Percentage of Players'].map('{:,.2%}'.format)
ageDf = ageDf.sort_index()
ageDf
```

## Age Analysis
- The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.):
  - Purchase Count
  - Average Purchase Price
  - Total Purchase Value
  - Average Purchase Total per Person by Age Group

```python
#set up the dataframe 
purchaseDf['Age Ranges'] = pd.cut(purchaseDf['Age'], ageBins, labels=ageLabels)

#calculate the purchase count
agePurchaseCount = purchaseDf.groupby(['Age Ranges']).count()['Price'].rename('Purchase Count')

#calculate the total purchase value
agePurchaseTotal = purchaseDf.groupby(['Age Ranges']).sum()['Price'].rename('Total Purchase Value')

#calculate the average purchase price
agePurchaseAvg = purchaseDf.groupby(['Age Ranges']).mean()['Price'].rename('Average Purchase Price')

#calculate the average purchase total per person by gender
agePurchaseAvgTotal = agePurchaseTotal/ageDf['Total Count']

#display the results in a table 
agePurchaseData = pd.DataFrame({'Purchase Count': agePurchaseCount,
                             'Average Purchase Price': agePurchaseAvg,
                             'Total Purchase Value': agePurchaseTotal,
                             'Average Purchase Total': agePurchaseAvgTotal
                            })
agePurchaseData = agePurchaseData.sort_index()
agePurchaseData['Average Purchase Price'] = agePurchaseData['Average Purchase Price'].map('${:,.2f}'.format)
agePurchaseData['Total Purchase Value'] = agePurchaseData['Total Purchase Value'].map('${:,.2f}'.format)
agePurchaseData['Average Purchase Total'] = agePurchaseData['Average Purchase Total'].map('${:,.2f}'.format)
agePurchaseData
```

## Top Spenders
- Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
  - SN
  - Purchase Count
  - Average Purchase Price
  - Total Purchase Value

```python
#identify the top 5 spenders
topSpenders = purchaseDf.groupby(['SN'])
topSpenders

#calculate the top 5's purchase count
topSpendersPC = topSpenders['SN'].count() 

#calculate the top 5's average purchase price
topSpendersAvg = topSpenders['Price'].mean()

#calculate the top 5's total purchase value
topSpendersTotal = topSpenders['Price'].sum()

#display the results in a table
spendersDf = pd.DataFrame({'Purchase Count': topSpendersPC,
        'Average Purchase Price':topSpendersAvg,
        'Total Purchase Value':topSpendersTotal})
spendersDf = spendersDf.sort_values(['Total Purchase Value'], ascending=False).head(5)
spendersDf['Average Purchase Price'] = spendersDf['Average Purchase Price'].map('${:,.2f}'.format)
spendersDf['Total Purchase Value'] = spendersDf['Total Purchase Value'].map('${:,.2f}'.format)
spendersDf
```

## Most Popular Items

Identify the 5 most popular items by purchase count, then list (in a table):

- Item ID
- Item Name
- Purchase Count
- Item Price
- Total Purchase Value

```python
#identify the top 5 items
topItems = purchaseDf.groupby(['Item ID', 'Item Name'])

#calculate the top 5's purchase count
topItemsPC = topItems['Item ID'].count() 

#identify the top 5's item price
topItemsPrice = topItems['Price'].mean()

#calculate the top 5's total purchase value
topItemsTotal = topItems['Price'].sum()

# #display the results in a table
topItems5 = pd.DataFrame({
                         'Purchase Count':topItemsPC,
                         'Item Price':topItemsPrice,
                         'Total Purchase Value':topItemsTotal})
topItems5 = topItems5.sort_values(['Purchase Count'], ascending=False).head(5)

topItems5['Total Purchase Value'] = topItems5['Total Purchase Value'].map('${:,.2f}'.format)
topItems5['Item Price'] = topItems5['Item Price'].map('${:,.2f}'.format)
topItems5
```

## Most Profitable Items
Identify the 5 most profitable items by total purchase value, then list (in a table):
- Item ID
- Item Name
- Purchase Count
- Item Price
- Total Purchase Value

```python
# #display the results in a table
topProfit5 = pd.DataFrame({
                         'Purchase Count':topItemsPC,
                         'Item Price':topItemsPrice,
                         'Total Purchase Value':topItemsTotal})
topProfit5 = topProfit5.sort_values(['Total Purchase Value'], ascending=False).head(5)

topProfit5['Total Purchase Value'] = topProfit5['Total Purchase Value'].map('${:,.2f}'.format)
topProfit5['Item Price'] = topProfit5['Item Price'].map('${:,.2f}'.format)
topProfit5
```
